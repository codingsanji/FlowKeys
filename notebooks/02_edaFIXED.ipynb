```python
df_c = df.drop(columns=['participant', 'session', 'repetition', 'total time'], axis=1)
corr = df_c.corr()

plt.figure(figsize=(32, 24))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix of Keystroke Features")
```
